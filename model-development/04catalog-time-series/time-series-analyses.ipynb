```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.path import Path
from shapely.geometry import Polygon, Point
from scipy.stats import gaussian_kde
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from pyslabgrid import eqcatana as eqc
```

```python
# catalogues
cat_folder = '../02midslab-config/projected_catalogs/'
# lon,lat, dep, mag, year, slat, slon, sdep, dproj
```

```python
catalog = 'NZeqcat_Rollins13042020-subd-slab-proj2midslab.csv'
eqcat = eqc.read_projectedslabcatalog(cat_folder+catalog, min_year=1900, min_mag=3.0)
slon = np.array(eqcat['slon'])
slat = np.array(eqcat['slat'])
sdep = np.array(eqcat['sdep'])
mag = np.array(eqcat['mag'])
year = np.array(eqcat['year'])

slabzone = 'hik-whole'
in_hikw = eqc.get_inpolgyon((eqcat['slon'],eqcat['slat']), eqc.get_zonepolygon(slabzone))
mag_hikw = mag[in_hikw]
year_hikw = year[in_hikw]

slabzone = 'puy-whole'
in_puyw = eqc.get_inpolgyon((eqcat['slon'],eqcat['slat']), eqc.get_zonepolygon(slabzone))
mag_puyw = mag[in_puyw]
year_puyw = year[in_puyw]

plt.rcParams.update({'font.size': 16})

fig, ax = plt.subplots(4,1, figsize=(13,10), sharex=True)

eqc.plot_yearlycounts(year_hikw, mag_hikw, ax[0])
ax[0].set_xlim([1900, 2020])
ax[0].set_ylim([1, 10e3])
ax[0].text(1888, 3800, '(a)', fontsize=22)
ax[0].legend(fontsize=11)
eqc.scatter_magdensity(year_hikw, mag_hikw, ax = ax[1], fig=fig)
ax[1].set_xlim([1900, 2020])
ax[1].set_ylim([3.0, 7.5])
ytics = [x for x in range(3, 8, 1)]
ax[1].set_yticks(ytics)
xtics = [x for x in range(1900, 2020, 5)]
ax[1].set_xticks(xtics)
ax[1].set_xticklabels(ax[1].get_xticks(),rotation = 45);


eqc.plot_yearlycounts(year_puyw, mag_puyw, ax[2])
ax[2].set_xlim([1900, 2020])
ax[2].set_ylim([1, 10e3])
ax[2].text(1888, 3800, '(b)', fontsize=22)
ax[2].legend(fontsize=11)
eqc.scatter_magdensity(year_puyw, mag_puyw, ax = ax[3], fig=fig)
ax[3].set_xlim([1900, 2020])
ax[3].set_ylim([3.0, 7.5])
ytics = [x for x in range(3, 8, 1)]
ax[3].set_yticks(ytics)
xtics = [x for x in range(1900, 2020, 5)]
ax[3].set_xticks(xtics)
ax[3].set_xticklabels(ax[1].get_xticks(),rotation = 45);

print('Orginal catalogue')

fig.savefig('Figure4.jpeg', dpi=300, facecolor='w')
```
