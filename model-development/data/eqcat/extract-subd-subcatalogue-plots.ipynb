```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

#### Plot the catalogues
The files are:
- NZeqcat_Rollins13042020-subd.csv
- Grue_NZeqcat_Rollins13042020SepiDeep-subd.csv
- Grue_NZeqcat_Rollins13042020SepiShallow_subd.csv


```python
def read_catalogue(input_file):
    # year, month, day, hour, minute, second, lon, lat, depth, mag
    eqcat = {}
    with open(input_file, "r") as f:
        datareader=csv.reader(f)
        header = next(datareader)
        nrow = 0
        year, lon, lat, dep, mag = [],[],[],[],[]
        for row in datareader:    
            year.append(int(row[0]))
            lon.append(float(row[6]))
            lat.append(float(row[7]))
            dep.append(float(row[8]))
            mag.append(float(row[9]))
        eqcat.update({'year': year, 'lon': lon, 
                  'lat':lat, 'dep':dep,
                  'mag':mag})
        return eqcat
```

```python
%matplotlib inline
catalogues = ['NZeqcat_Rollins13042020-subd.csv',
    'Grue_NZeqcat_Rollins13042020SepiDeep-subd.csv',
    'Grue_NZeqcat_Rollins13042020SepiShallow_subd.csv',]
folder = 'subduction_catalogs/'
for catalog in catalogues: 
    eqcat = read_catalogue(folder+catalog)
    plt.scatter(eqcat['lon'], eqcat['lat'], c= eqcat['dep'])
    plt.show()
```

```python
%matplotlib notebook
catalogues = ['NZeqcat_Rollins13042020-subd.csv',]

fig = plt.figure(figsize=(7,7))

k=1
for catalog in catalogues: 
    eqcat = read_catalogue(folder+catalog)
    ax = fig.add_subplot(1,1,k, projection='3d')
    ax.scatter(eqcat['lon'], eqcat['lat'], eqcat['dep'], c=eqcat['dep'])
    ax.set_xlabel("lon")
    ax.set_ylabel("lat")
    ax.set_zlabel("dep")
    ax.invert_zaxis()
    k+=1

```
