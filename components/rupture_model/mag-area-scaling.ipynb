```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def AllenHayes2017(mag):
    return 10**(-3.89 + 0.96 * mag)
def StrasserEtAl2010(mag):
    return 10 **(-3.225 + 0.89 * mag)
def Iwata_Asano2011(mag):
    return 10**(-4.149+mag)
def cscaling(mag, c=4.2):
    return 10**(-c+mag)
```

```python
M = [m for m in np.arange(5.0, 9.1, 0.1)]

scalerel = ['Allen and Hayes (2017)', 'Strasser et al. (2010)', \
            'Iwata and Asano (2011)', 'self-similar, c=4.2']

mycol = ['b', 'r', 'g', 'm']
plt.figure(figsize=(6, 5), dpi=80)
plt.rcParams.update({'font.size': 12})
i=0
for rel in scalerel:
    if rel == 'Allen and Hayes (2017)':
        A = [AllenHayes2017(mag) for mag in M]
    if rel == 'Strasser et al. (2010)':
        A = [StrasserEtAl2010(mag) for mag in M]
    if rel == 'Iwata and Asano (2011)':
        A = [Iwata_Asano2011(mag) for mag in M]
    if rel == 'self-similar, c=4.2':
        A = [cscaling(mag, c=4.2) for mag in M]
    
    plt.semilogy(M, A, '-', label=rel, color=mycol[i])
    i+=1

plt.legend()
plt.xlabel('Magnitude, Mw')
plt.ylabel('Rupture Area (km$^2$)');
plt.xlim([5, 9]);
```
