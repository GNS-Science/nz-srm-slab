```python
def read_ratefile(infile):
    x,y,z,r = [],[],[],[]
    with open(infile, 'r') as f:
        csvreader = csv.reader(f)
        header = next(csvreader)
        for row in csvreader:
            x.append(float(row[0]))
            y.append(float(row[1]))
            z.append(float(row[2]))  
            r.append(float(row[3]))
    return x,y,z,r 
```

```python
import nzborders as nzb

outfolder = 'files/processed-smooth/'
files = ['hik-slab-smooth_1depth.csv', 'hik-slab-smooth_lay1.csv',
         'hik-slab-smooth_lay2.csv', 'hik-slab-smooth_lay3.csv',]
for file in files:
    x,y,z, r = read_ratefile(outfolder+file)
    print(sum(r))
    fig, ax = plt.subplots(figsize=(10,7))
    print(file, 'min rate=', min(r), 'max rate =', max(r))
    sc = ax.scatter(x,y, c=r, cmap='plasma_r')
    nzb.plot(file='nzborders.json', plinewidth= 0.2, pcolor = 'k', ax=ax)
    fig.colorbar(sc, ax = ax, shrink=0.5);
    plt.show()
```
